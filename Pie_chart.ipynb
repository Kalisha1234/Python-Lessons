```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
# defining proteins and sizes
labels = ['Meat', 'Beef', 'Sausage', 'Egg', 'FIsh']
size = [20, 15, 15, 40, 10]
```

```python
plt.figure(figsize=(5,5))
plt.pie(size, labels=labels, autopct='%1.1f%%')
plt.title('Proteins')
plt.show()
```
